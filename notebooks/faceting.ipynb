### Challenge 2

For this challenge you will be working with the **cars** data set. Run the cell below to import the data set.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cars_data = pd.read_csv('../data/cars.csv', sep='|')
cars_data.rename(columns={'Unnamed: 0': 'make_model'}, inplace=True)
cars_data.head()
````

For this challenge:
* Facet the plots on the number of cylenders each vechical has
* Create a single row with with the number of correct plots
* Use a scatter plot to show the relation between horse power and quarter mile time
* On the x axis plot the horse power
* On the y axis plot the the quarter mile time (qsec)
* Each plot should have a title saying the number of cylinders
* Give each axis an appropriate name

````python
# Write your code here
plt.figure()
plt.figtext(0,1, "Relationship between hp and qsec")

grouped = cars_data.groupby('cyl')
four = grouped.get_group(4)
six = grouped.get_group(6)
eight = grouped.get_group(8)


#first subplot
plt.subplot(1,3,1)
plt.title("four cylinder")
plt.xlabel("hp")
plt.ylabel("qsec")
plt.scatter(four['hp'],four['qsec'])

#second subplot
plt.subplot(1,3,2)
plt.title("six cylinder")
plt.xlabel("hp")
plt.ylabel("qsec")
plt.scatter(six['hp'],six['qsec'])

#third subplot
plt.subplot(1,3,3)
plt.title("eight cylinder")
plt.xlabel("hp")
plt.ylabel("qsec")
plt.scatter(eight['hp'],eight['qsec'])

plt.tight_layout()
plt.show()
````
